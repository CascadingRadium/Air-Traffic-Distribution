```python
import pandas as pd
df=pd.read_csv("mutation.csv")
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X=range(0,251)
fig=plt.figure(figsize=(15,15))
ax = plt.gca()
ax.set_xlim([0, 255])
ax.set_ylim([0, 0.000040])
L=[]
for index,val in enumerate(list(df.columns)):
    plt.plot(X,df.iloc[:, index].tolist(),label=val)
    L.append(max(df.iloc[:, index].tolist()))
plt.legend()
```

```python
plt.plot(list(df.columns),L)
```
