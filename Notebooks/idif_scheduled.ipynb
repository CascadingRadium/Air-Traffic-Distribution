```python
import pandas as pd
```

```python
final_df = pd.read_csv("2013-08-16_flights.csv")
```

```python
from datetime import datetime
min_date = min(final_df['actual_runway_departure'])
min_date = datetime.strptime(min_date, '%Y-%m-%d %H:%M:%S')
print(min_date)
```

```python
l4 = []
for index, row in final_df.iterrows():
    dt_obj = datetime.strptime(row['scheduled_runway_departure'], '%Y-%m-%d %H:%M:%S')
    l4.append(dt_obj)
final_df['dt_scheduled_runway_departure'] = l4
```

```python
l2 = []
for index, row in final_df.iterrows():
    dif = row['dt_scheduled_runway_departure'] - min_date
    l2.append(dif)
final_df['dif_scheduled_runway_departure'] = l2
```

```python
l3 = []
min_timestamp = int(round(min_date.timestamp()))
for index, row in final_df.iterrows():
    idif = int((int(round(row['dt_scheduled_runway_departure'].timestamp())) - min_timestamp)/60)
    idif -= 330
    l3.append(idif)
final_df['idif_scheduled_runway_departure'] = l3
```

```python
final_df.drop("dt_scheduled_runway_departure",axis=1,inplace=True)
final_df.drop("dif_scheduled_runway_departure",axis=1,inplace=True)
```
